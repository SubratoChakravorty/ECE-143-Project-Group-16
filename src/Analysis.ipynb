## Teacher Attributes

We then look at the influence of all teaching variables like teacher’s use of group work, introduction of new ideas, clarity of concepts etc., and observe something extremely interesting. It turns out that the science self-efficacy of a student is majorly influenced by just one factor, the affirmative attitude. 
Affirmative attitude is computed as a combination of respect towards student ideas, appreciation of student’s work, differential treatment for male and female students, mistakes are OK, treats every student fairly, and instills confidence in students. Each of these were scored between 1 and 4 by students in a questionnaire and we combine these variables by taking the average across them all. 1 represents strongly negative and 4 represents strongly positive reaction to these. 





```python
linear_regression_affatt(st)
```

```python
affatt_boxplots(st)
```
